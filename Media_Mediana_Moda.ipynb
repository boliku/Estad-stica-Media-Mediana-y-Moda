# Estad√≠stica: Media, Mediana y Moda

## Media

La media es la suma del conjunto de valores dividida entre el n√∫mero total de valores

![Media](./Images/media_arit.png)


### Promedio de edades

```python
edades = [20, 15, 41, 19, 62, 37, 14]
promedio = sum(edades)/len(edades)
print("Con funciones nativas:", promedio)


import numpy as np
edades = np.array([20, 15, 41, 19, 62, 37, 14])
promedio = edades.mean()
print("Con numpy:", promedio)

# Redondeo
print("Redondeado:", round(promedio), "a√±os")
```

# Moda
La moda es el valor que aparece con mayor frecuencia en un conjunto de valores de datos.

![Moda.pnh](./Images/moda.png)

```python
notas = [12, 20, 15, 18, 16, 15, 12, 15, 11, 15, 12, 18]
cantidad = []
for i in notas:
    cantidad.append(notas.count(i))
moda = notas[cantidad.index(max(cantidad))]
print("La moda es:", moda)
```

# Mediana

La mediana es el valor que ocupa el lugar central de todos los datos cuando √©stos est√°n ordenados de menor a mayor.

![Mediana.png](./Images/mediana.png)

### Mediana de edades

```python
edades = [20, 15, 41, 19, 62, 37, 14, 1]
edades = sorted(edades)
print("Edades ordenadas:", edades)

if len(edades) % 2 != 0:
    mediana = edades[(len(edades)-1)//2]
else:
    mediana = (edades[len(edades)//2-1] + edades[len(edades)//2])/2
print("Con funciones nativas:", mediana)


import numpy as np
edades = np.array([20, 15, 41, 19, 62, 37, 14, 1])
mediana = np.median(edades)
print("Con numpy:", mediana)
```
